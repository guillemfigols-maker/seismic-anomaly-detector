## Data ingestion from USGS  

```python
# imports
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
# we first want to fetch all the data from the api
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
chunk_days = 30
start_date = "2024-01-01"
end_date = "2025-01-01"
current = datetime.strptime(start_date, "%Y-%m-%d")
end = datetime.strptime(end_date, "%Y-%m-%d")
data = []
while current < end:
    chunk_end = min(current + timedelta(days=chunk_days), 
                    end
                )
    params = {
        "format": "geojson",
        "starttime": current.strftime("%Y-%m-%d"),
        "endtime": chunk_end.strftime("%Y-%m-%d")
    }
    response = requests.get(url, params=params)
    response.raise_for_status()
    data_url = response.json()
    data.extend(data_url["features"])
    current = chunk_end

```

```python
df_eq = pd.DataFrame(data)
```
